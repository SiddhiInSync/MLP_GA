<h2> Common Instructions for Question 6 and 7

Rules for Imputing the missing(NaN) or Unknown values:

Calculating statistical values (such as mean, median, mode) for each column in the training dataset.

Applying these calculated statistical values to replace missing (NaN) and unknown values in both the training and test datasets.

Ensure that the calculation of statistical values excludes any rows containing missing or unknown values.

Replace Unknown values in the "Age" feature with the Mean value in that. Replace Unknown values in the "Location" feature with the constant value "Other".

Replace Unknown values in the "GameDifficulty" feature with the Most Frequent value in that.

Replace Unknown values in the "InGamePurchases" feature with the constant value '0'.

Write the answers related to the above imputation in below questions respectively.

```python
#Calculating statistical values (such as mean, median, mode) for each column in the training dataset.

mean_values = X_train.mean(numeric_only = True)
median_values = X_train.median(numeric_only = True)
mode_values = X_train.mode().iloc[0]
```

```python
# Applying these calculated statistical values to replace missing (NaN) and unknown values in both the training and test datasets.
# Ensure that the calculation of statistical values excludes any rows containing missing or unknown values.

X_train.isnull().sum()
```

```python
# Replace Unknown values in the "Age" feature with the Mean value in that.
X_train['Age'] = X_train['Age'].fillna(mean_values['Age'])
X_test['Age'] = X_test['Age'].fillna(mean_values['Age'])
```

```python
X_train['Location'] = X_train['Location'].fillna('Other')
X_test['Location'] = X_test['Location'].fillna('Other')
```

```python
from sklearn.impute import SimpleImputer
```

```python
X_train['GameDifficulty'] = X_train['GameDifficulty'].fillna(mode_values['GameDifficulty'])
X_test['GameDifficulty'] = X_test['GameDifficulty'].fillna(mode_values['GameDifficulty'])
```

```python
X_train['InGamePurchases'] = X_train['InGamePurchases'].fillna(0)
X_test['InGamePurchases'] = X_test['InGamePurchases'].fillna(0)
```

<h2> Ques. 7

```python
#Drop PlayerId
X_train.drop(columns=['PlayerID'], inplace=True)
X_test.drop(columns=['PlayerID'], inplace=True)
```

```python
#Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()
X_train['GameDifficulty'] = oe.fit_transform(X_train[['GameDifficulty']])
X_test['GameDifficulty'] = oe.transform(X_test[['GameDifficulty']])
```

```python
#one hot encoding

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')
train_ohe = ohe.fit_transform(X_train[['Gender', 'Location', 'GameGenre']])
test_ohe = ohe.transform(X_test[['Gender', 'Location', 'GameGenre']])
```

```python
#Remove Categorical columns

X_train_num = X_train.drop(columns=['Gender', 'Location', 'GameGenre'])
X_test_num = X_test.drop(columns=['Gender', 'Location', 'GameGenre'])
```

```python
# OneHotEncoder returns Numpy not a DataFrame
# so , we have to join OHE columns

ohe_columns = ohe.get_feature_names_out(['Gender', 'Location', 'GameGenre'])
X_train_final  = pd.concat([X_train_num.reset_index(drop=True), pd.DataFrame(train_ohe, columns = ohe_columns)], axis=1)
X_test_final = pd.concat([X_test_num.reset_index(drop=True), pd.DataFrame(test_ohe, columns= ohe_columns)], axis=1)
```

```python
#Standard Scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().set_output(transform='pandas')
X_train_scaled = scaler.fit_transform(X_train_final)
X_test_scaled = scaler.transform(X_test_final)
```
